```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
import re
load_dotenv()
import time
from bs4 import BeautifulSoup as bs
```

```python
KEY=os.getenv("scm_api")
TYPE="json"
SERVICE="VwsmAdstrdNcmCnsmpW"
START_INDEX=1
END_INDEX=10
url = f"http://openapi.seoul.go.kr:8088/{KEY}/{TYPE}/{SERVICE}/{START_INDEX}/{END_INDEX}"
r=requests.get(url)
response=r.json()
response
```

```python
response['VwsmAdstrdNcmCnsmpW']['list_total_count']
```

```python
response['VwsmAdstrdNcmCnsmpW']['row']
```

```python
start=1
end=1000
page_num=1
total_page=1
result={}

while page_num<=total_page:
    KEY=os.getenv("scm_api")
    TYPE="json"
    SERVICE="VwsmAdstrdNcmCnsmpW"
    START_INDEX=start
    END_INDEX=end
    url = f"http://openapi.seoul.go.kr:8088/{KEY}/{TYPE}/{SERVICE}/{START_INDEX}/{END_INDEX}/20220301"
    r=requests.get(url)
    response=r.json()
    
    if total_page==1:
        total_page=int(response['VwsmAdstrdNcmCnsmpW']['list_total_count'])//1000+1
    else:
        pass
    
    for item in response['VwsmAdstrdNcmCnsmpW']['row']:
        for key,value in item.items():
            if type(value)==float:
                result.setdefault(key,[]).append(int(value))
            else:
                result.setdefault(key,[]).append(value)
                
    print(f"{page_num}/{total_page} 수집중입니다.", end="\r")        
    page_num+=1
    start+=1000
    end+=1000
    time.sleep(0.2)

df_json=pd.DataFrame(result)
df_json
    
```

```python
df_json.to_csv("./data/json으로불러온상권분석데이터.csv",index=False,encoding="utf-8-sig")
```

```python
start=1
end=1000
page_num=1
total_page=1
result={}

while page_num<=total_page:
    KEY=os.getenv("scm_api")
    TYPE="xml"
    SERVICE="VwsmAdstrdNcmCnsmpW"
    START_INDEX=start
    END_INDEX=end
    url = f"http://openapi.seoul.go.kr:8088/{KEY}/{TYPE}/{SERVICE}/{START_INDEX}/{END_INDEX}/20220301"
    r=requests.get(url)
    response=r.text
    soup=bs(response,'xml')
    
    if total_page==1:
        total_page=int(soup.select_one('list_total_count').text)//1000+1
    else:
        pass
    
    
    cols=['STDR_YYQU_CD','ADSTRD_CD','ADSTRD_CD_NM','MT_AVRG_INCOME_AMT','INCOME_SCTN_CD','EXPNDTR_TOTAMT','FDSTFFS_EXPNDTR_TOTAMT','CLTHS_FTWR_EXPNDTR_TOTAMT','LVSPL_EXPNDTR_TOTAMT',
      'MCP_EXPNDTR_TOTAMT','TRNSPORT_EXPNDTR_TOTAMT','EDC_EXPNDTR_TOTAMT','PLESR_EXPNDTR_TOTAMT','LSR_CLTUR_EXPNDTR_TOTAMT','ETC_EXPNDTR_TOTAMT','FD_EXPNDTR_TOTAMT']
    kor_cols=["기준_년분기_코드","행정동_코드","행정동_코드_명","월_평균_소득_금액","소득_구간_코드","지출_총금액","식료품_지출_총금액","의류_신발_지출_총금액",
              "생활용품_지출_총금액","의료비_지출_총금액","교통_지출_총금액","교육_지출_총금액","유흥_지출_총금액","여가_문화_지출_총금액","기타_지출_총금액","음식_지출_총금액"]
    
    for key, kor_col in zip(cols,kor_cols):
        for value in soup.select(key):
            result.setdefault(kor_col,[]).append(value.text)
            
    print(f"{page_num}/{total_page} 수집중입니다.",end="\r")        
    start+=1000
    end+=1000
    page_num+=1
    time.sleep(0.2)
    

df_xml=pd.DataFrame(result)
df_xml
```
