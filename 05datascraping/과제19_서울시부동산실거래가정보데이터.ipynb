```python
from dbio import to_db
```

```python
start = 1
end = 1000
page = 1
total_page = 1
result = {}

while page <= total_page:
    url = f"http://openapi.seoul.go.kr:8088/{os.getenv('seoul_api')}/json/tbLnOpendataRtmsV/1/5/"
    r = requests.get(url)
    print(r.status_code)
    response = r.json()
    if total_page == 1:
        total_page = response['tbLnOpendataRtmsV']['list_total_count'] // 1000 + 1


    for item in response['tbLnOpendataRtmsV']['row']:
        for key, value in item.items():
            if type(value) == float:
                result.setdefault(key, []).append(int(value))
            else:
                result.setdefault(key, []).append(value)

    print(f"{page}/{total_page} 데이터 수집중", end="\r")            
    start += 1000
    end += 1000
    page += 1
    
    time.sleep(0.2)
            
df = pd.DataFrame(result)
to_db("seoul_open_api","서울시_부동산_실거래가",df)
```
