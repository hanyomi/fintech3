```python
hotel_id=[]
region=[]
locality=[]
name=[]
star_rating=[]
rating=[]
total_count=[]
rating_aspects=[]
for i in range(0,len(data)):
    if not data[i]['data']['body']['pdpHeader']['hotelId'] in hotel_id:
        hotel_id.append(data[i]['data']['body']['pdpHeader']['hotelId'])
        region.append(data[i]['data']['body']['propertyDescription']['address'].get('region', None))
        locality.append(data[i]['data']['body']['propertyDescription']['address']['locality'])
        name.append(data[i]['data']['body']['propertyDescription']['name'])
        star_rating.append(data[i]['data']['body']['propertyDescription']['starRating'])
        rating.append(data[i]['data']['body']['reviewContent']['overall']['rating'])
        total_count.append(data[i]['data']['body']['reviewContent']['overall']['totalCount'])
        rating_aspects.append(data[i]['data']['body']['reviewContent']['overall']['ratingAspects'])

    
print(hotel_id,region,locality,name,star_rating,rating)
```

```python
result={"hotel_id":hotel_id,"name":name,"star_rating":star_rating,"total_count":total_count,"rating":rating,"rating_aspects":rating_aspects,"region":region,"locality":locality}
result
```

# 호텔별 사용자 리뷰 추출

```python
hotel_reviews={}
for index,hotel in enumerate(data):
    print(f"{index}/{len(data)}",end="\r")
    hotel_id=hotel['data']['body']['pdpHeader']['hotelId']
    try:
        reviews=hotel['data']['body']['reviewContent']['reviews']['hermes']['groups'][0]['items']
    except Exception as e:
        print(e)
        print("이 호텔은 리뷰가 없습니다.")
        
    
    for review in reviews:
        user_id=review['itineraryId']
        trip_type=review['tripType']
        review_date=review['reviewDate']
        u_rating=review['rating']
        review_text=review['description']

        hotel_reviews.setdefault("hotel_id",[]).append(hotel_id)
        hotel_reviews.setdefault("user_id",[]).append(user_id)
        hotel_reviews.setdefault("trip_type",[]).append(trip_type)
        hotel_reviews.setdefault("review_date",[]).append(review_date)
        hotel_reviews.setdefault("u_rating",[]).append(u_rating)
        hotel_reviews.setdefault("review_text",[]).append(review_text)
hotel_reviews_df=pd.DataFrame(hotel_reviews)
hotel_reviews_df
```
