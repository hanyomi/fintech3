## db 저장하는 함수 만들기

```python
engine = create_engine("mysql+pymysql://root:1234@localhost:3306")
conn = engine.connect()
conn.execute(text("create database if not exists fintech_news"))
conn.close()
```

```python
def fintech_news(page, start_date, end_date):

    page_result = []

    url = "https://www.fintech.or.kr/web/board/boardContentsList.do"
    payload = dict(board_id=6, mode='W', miv_pageNo=page)
    r = requests.post(url, params=payload)
    response = bs(r.content, 'lxml')
    notice_list = response.select("tbody>tr")

    for notice in notice_list:
        # 날짜 가져오기
        notice_date = notice.select_one("td.last").text.strip()
        if not notice_date:
            continue

        # 날짜 포맷 변환
        if "/" in notice_date:
            month, day = map(int, notice_date.split("/"))
            notice_date = datetime(2025, month, day).date()
        else:
            notice_date = datetime.strptime(notice_date, "%Y-%m-%d").date()

        # 날짜 범위 벗어나면 스킵
        if not (end_date <= notice_date <= start_date):
            continue

        # '핀테크 주요뉴스' 포함 제목만
        notice_title = notice.select_one("td.txtl.title a").text.strip()
        if "핀테크 주요뉴스" not in notice_title:
            continue

        # 뉴스 링크
        a_tag = notice.select_one("td.txtl.title a")
        link = a_tag["href"]
        id_link = link.split("(")[1].split(")")[0].strip("'")

        # 뉴스 리스트 가져오기
        news_url = "https://www.fintech.or.kr/web/board/boardContentsView.do"
        payload = dict(board_id=6, mode='W', miv_pageNo=1, contents_id=id_link, searchkey='I')
        r = requests.post(news_url, params=payload)
        n_response = bs(r.content, 'lxml')

        news_list = n_response.select("table tr a")
        for lists in news_list[:-4]:  # 마지막 4개 제외
            news_title = lists.get_text(strip=True)
            news_link = lists["href"]

            if "?" in news_link:
                news_url_base = news_link.split("?")[0]
                news_param = news_link.split("?")[1]
            else:
                news_url_base = news_link
                news_param = ""

            r = requests.get(news_url_base, params=news_param)
            d_response = bs(r.content, 'lxml')

            # 뉴스 본문 가져오기
            if "https://n.news.naver.com/mnews/article" in news_url_base:
                article = d_response.select_one('article#dic_area')
                desc = article.get_text(separator="\n", strip=True) if article else ""
            else:
                paragraphs = d_response.select("p")[:-3]  # 마지막 3개 제외
                desc = " ".join([p.get_text(strip=True) for p in paragraphs])

            page_result.append((notice_date, news_title, news_link, desc))

    return page_result
```

```python
start_date = datetime(2025, 11, 20).date()
end_date = datetime(2025, 1,2).date()
all_results = []

for page in range(1, 20):
    print(f"{page}/19 데이터 수집중", end="\r")
    try:
        page_result = fintech_news(page, start_date, end_date)
        all_results.extend(page_result)
    except Exception as e:
        print(e)

# DataFrame 생성
cols = ['date', 'title', 'link', 'content']
df = pd.DataFrame(all_results, columns=cols)

# MySQL에 한 번만 저장
engine = create_engine("mysql+pymysql://root:1234@localhost:3306/fintech_news")
df.to_sql("fintech_news", con=engine, if_exists="replace", index=False)
print("\n데이터 저장 완료!")
```
